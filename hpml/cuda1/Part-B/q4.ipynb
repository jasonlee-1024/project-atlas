```python

import matplotlib.pyplot as plt
import numpy as np

K = [1, 5, 10, 50, 100]

# CPU only (ms -> sec)
cpu_ms = [0.000, 0.003, 0.006, 0.034, 0.067]
cpu_sec = [t / 1000 for t in cpu_ms]

# GPU Scenarios (sec)
s1 = [0.038265, 0.315325, 0.631571, 3.154956, 6.334772]
s2 = [0.000617, 0.007236, 0.014406, 0.071942, 0.143587]
s3 = [0.000021, 0.000245, 0.000476, 0.002495, 0.004998]

plt.figure(figsize=(10, 6))
plt.plot(K, cpu_sec, 'r--D', label='CPU only', markersize=7)
plt.plot(K, s1, 'b-o', label='Scenario 1', markersize=7)
plt.plot(K, s2, 'g-s', label='Scenario 2', markersize=7)
plt.plot(K, s3, 'm-^', label='Scenario 3', markersize=7)

plt.xscale('log')
plt.yscale('log')
plt.xlabel('K (million elements)')
plt.ylabel('Execution time (sec, log scale)')
plt.title('CPU vs GPU Execution Time')
plt.legend()
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.xticks(K, [str(k) for k in K])
plt.tight_layout()
plt.savefig('q4_with_unified.jpg', dpi=150)
plt.show()
```
